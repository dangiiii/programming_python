```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('../data/fandango_score_comparison.csv')
```

```python
# Choosing an appropriate colormap
data.plot.scatter(y='RottenTomatoes', x='IMDB_user_vote_count', c='RottenTomatoes_User', vmin=0, vmax=100,
                  edgecolor='None', cmap='inferno')
plt.axhline(50, color='k', linestyle='--')
plt.ylim(0,100)
```

```python
#Selecting colors from a colormap
cm=plt.cm.get_cmap('cool')
for i in range(20):
    plt.plot(np.arange(0,10,0.1), np.sin(i*np.pi/20.+np.arange(0,10,0.1)), color=cm(i/20.))
```
