```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('../data/fandango_score_comparison.csv')
```

```python
data.head()
```

```python
#Distribution plot
sns.distplot(data['RT_norm'])
```

```python
#KDE Plot
sns.kdeplot(data['RT_norm'], shade=True)
```

```python
#2D KDE Plot
sns.kdeplot(data['RT_norm'], data['RT_user_norm'], shade=True)
```

```python
#The Joint plot
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='hex')
```

```python
#Theming Seaborn
sns.set_style('ticks')
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
```

```python
#Spine Control
sns.set_style('whitegrid')
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
sns.despine(left=True, bottom=True)
```

```python
#Automatic Scaling
sns.set_style('whitegrid')
sns.set_context('poster')
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
```
