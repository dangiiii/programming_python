```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('../data/fandango_score_comparison.csv')
```

```python
data.head()
```

```python
scores=['RT_norm', 'IMDB_norm', 'RT_user_norm']
```

```python
#Bar Chart with Pandas
data.head().plot.bar(x='FILM', y=scores)
plt.ylabel('score')
```

```python
#Horizontal Bars
data.head().plot.barh(x='FILM', y=scores)
plt.xlabel('score')
```

```python
#Bar Colors
data.head().plot.bar(x='FILM', y=scores, color=['y','r','k'])
plt.ylabel('score')
```

```python
#Bar Hatching
data.head().plot.bar(x='FILM', y=scores, hatch='x')
plt.ylabel('score')
```

```python
#Bar Width
data.head().plot.bar(x='FILM', y=scores, width=1)
plt.ylabel('score')
```

```python
#Stacked Charts
data.head().plot.bar(x='FILM', y=['Metacritic_user_vote_count','Fandango_votes', 'IMDB_user_vote_count'],
                     stacked=True, log=True)
                     
plt.ylabel('score')
```
